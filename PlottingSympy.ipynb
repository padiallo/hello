```python
#math-with-python/asrarshare
#58

import sympy as symp

from sympy.abc import x

y =x**2

p= symp.plotting.plot(y,show=False) #
p.xlim=[0,50]
p[0].line_color='m' #proprité particulière donc je choisis un élément
p.title ='Je change une proprité globale qui touche le graphe entier'
p.show()
```

```python
from sympy import symbols
x,a= symbols('x,a')
expr = a/x
n= 5
m=2*n
p = symp.plotting.plot(expr.subs(a,n),(x,-5,5),show=False)
p[0].label ='$y=%s$' %symp.latex(expr.subs(a,n))
p2 = symp.plotting.plot(expr.subs(a,m),show=False) #Ma deuxieme fonction à afficher
p.extend(p2) #Extend signifie étendre le premier graphe pour intégrer une deuxieme fonction; possible de faire p.extend(expr.subs(a,m),show=False) 
p2[0].label ='$y=%s$' %symp.latex(expr.subs(a,m))
p.ylim=[-5,5]
p.legend=True
p.show()

```
